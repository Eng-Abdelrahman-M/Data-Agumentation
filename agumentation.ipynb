```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("data/10.jpg")
```

# Flip

```python
vertical_flip = cv2.flip(img,0) # Vertical flip
horizontal_flip = cv2.flip(img,1) # Horizontal flip
```

```python
fig, axes = plt.subplots(1,3,figsize=(30,30))
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title("Original Image",fontsize=30)

axes[1].imshow(vertical_flip)
axes[1].axis('off')
axes[1].set_title("Vertical flip",fontsize=30)

axes[2].imshow(horizontal_flip)
axes[2].axis('off')
axes[2].set_title("Horizontal flip",fontsize=30)
plt.show()
```

# Crop

```python
img.shape
```

```python
cropped_image = img[80:214, 150:330]
```

```python
fig, axes = plt.subplots(1,2,figsize=(30,30))
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title("Original Image",fontsize=30)

axes[1].imshow(cropped_image)
axes[1].axis('off')
axes[1].set_title("Cropped image",fontsize=30)
plt.show()
```

# Rotation

```python
rotate_90 = cv2.rotate(img,rotateCode=cv2.ROTATE_90_CLOCKWISE)
rotate_180 = cv2.rotate(img,rotateCode=cv2.ROTATE_180)
```

```python
fig, axes = plt.subplots(1,3,figsize=(30,30))
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title("Original Image",fontsize=30)

axes[1].imshow(rotate_90)
axes[1].axis('off')
axes[1].set_title("Rotate 90 degree",fontsize=30)

axes[2].imshow(rotate_180)
axes[2].axis('off')
axes[2].set_title("Rotate 180 degree",fontsize=30)
plt.show()
```

# Translation

```python
# Store height and width of the image
height, width = img.shape[:2]
transformation_matrix = np.float32([[1, 0, 100], [0, 1, 100]])

# We use warpAffine to transform
# the image using the matrix, T
img_translation = cv2.warpAffine(img, transformation_matrix, (width, height))
```

```python
fig, axes = plt.subplots(1,2,figsize=(30,30))
axes[0].imshow(img)
axes[0].axis('off')
axes[0].set_title("Original Image",fontsize=30)

axes[1].imshow(img_translation)
axes[1].axis('off')
axes[1].set_title("Cropped image",fontsize=30)
plt.show()
```
